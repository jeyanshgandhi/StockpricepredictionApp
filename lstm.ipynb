```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("stock.csv")
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
df['Ticker'].unique()
```

```python
company = ['HDFCBANK.NS', 'ICICIBANK.NS', 'AXISBANK.NS', 'INDUSINDBK.NS',
       'YESBANK.NS', 'FEDERALBNK.NS']
for i in company:
    if i == 'HDFCBANK.NS':
        train_hdfc, test_hdfc = train_test_split(df[df['Ticker'] == i], test_size=0.2, random_state=42)
        train_hdfc.sort_values(by='Date', inplace=True)
        test_hdfc.sort_values(by='Date', inplace=True)
    elif i == 'ICICIBANK.NS':
        train_icici, test_icici = train_test_split(df[df['Ticker'] == i], test_size=0.2, random_state=42)
        train_icici.sort_values(by='Date', inplace=True)
        test_icici.sort_values(by='Date', inplace=True)
    elif i == 'AXISBANK.NS':
        train_axis, test_axis = train_test_split(df[df['Ticker'] == i], test_size=0.2, random_state=42)
        train_axis.sort_values(by='Date', inplace=True)
        test_axis.sort_values(by='Date', inplace=True)
    elif i == 'INDUSINDBK.NS':
        train_indus, test_indus = train_test_split(df[df['Ticker'] == i], test_size=0.2, random_state=42)
        train_indus.sort_values(by='Date', inplace=True)
        test_indus.sort_values(by='Date', inplace=True)
    elif i == 'YESBANK.NS':
        train_yes, test_yes = train_test_split(df[df['Ticker'] == i], test_size=0.2, random_state=42)
        train_yes.sort_values(by='Date', inplace=True)
        test_yes.sort_values(by='Date', inplace=True)
    elif i == 'FEDERALBNK.NS':
        train_fed, test_fed = train_test_split(df[df['Ticker'] == i], test_size=0.2, random_state=42)
        train_fed.sort_values(by='Date', inplace=True)
        test_fed.sort_values(by='Date', inplace=True)
    else:
      print("Error occured!!")
```

```python
df_train = pd.concat([train_hdfc, train_icici, train_axis, train_indus, train_yes, train_fed])
df_train
```

```python
df_test = pd.concat([test_hdfc, test_icici, test_axis, test_indus, test_yes, test_fed])
df_test
```

```python
def create_sequence(data, features, target='Close', look_back=60):
    X, y = [], []
    for i in range(look_back, len(data)):
        X.append(data[features].iloc[i-look_back:i].values)
        y.append(data[target].iloc[i])

    return np.array(X), np.array(y)
```

```python
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA_50', 'RSI'] + list(new_train.columns[new_train.columns.str.contains("Ticker_")])
print(features)
X_train,y_train = create_sequence(new_train, features, target='Close', look_back=60)

print("x:", X_train.shape)
print("y:", y_train.shape)
```

```python
print(y_train, X_train)
```

```python
X_test, y_test = create_sequence(new_test, features, target='Close', look_back=60)
```

```python
x_train = X_train.reshape(X_train.shape[0], X_train.shape[1], len(features))
```

```python
x_test = X_test.reshape(X_test.shape[0], X_test.shape[1], len(features))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(units=50, return_sequences=True))
model.add(LSTM(units=50))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=64, verbose=1)
```

```python
loss = model.evaluate(X_test, y_test)
print("Test Loss:", loss)
```

```python
# Make predictions on the test set
predictions = model.predict(X_test)

# If you want to visualize or compare with actual values
import matplotlib.pyplot as plt

# Plotting the predictions against the actual values
plt.figure(figsize=(12, 6))
plt.plot(y_test, label='Actual Prices', color='blue')
plt.plot(predictions, label='Predicted Prices', color='orange')
plt.title('Actual vs Predicted Prices')
plt.xlabel('Time Steps')
plt.ylabel('Price')
plt.legend()
plt.show()
```
