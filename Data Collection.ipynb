```python
import pandas as pd
import yfinance as yf
```

```python
tickers = ['HDFCBANK.NS', 'ICICIBANK.NS', 'AXISBANK.NS', 'INDUSINDBK.NS', 'YESBANK.NS', 'FEDERALBNK.NS']
start_date = '2002-01-01'
end_date = '2024-10-04'

# Fetching data for all companies
data_dict = {}
for ticker in tickers:
    data_dict[ticker] = yf.download(ticker, start=start_date, end=end_date)
```

```python
# Adding ticker column and combining all stock data into one DataFrame
for ticker in tickers:
    data_dict[ticker]['Ticker'] = ticker  # Adding ticker column
combined_data = pd.concat([data_dict[ticker] for ticker in tickers], axis=0)
```

```python
combined_data
```

```python
combined_data.reset_index(inplace = True)
```

```python
combined_data.info()
```

```python
df = combined_data
```

```python
df
```

```python
df['MA_50'] = df.groupby('Ticker')['Close'].transform(lambda x: x.rolling(window=50).mean())
```

```python
df
```

```python
df['MA_50'].isna().sum()
```

```python
df.dropna(subset=['MA_50'], inplace=True)
```

```python
df
```

```python
df['MA_50'].isna().sum()
```

```python
# Parameters
window_length = 14  # RSI window length, typically 14

# Calculate the daily price change
delta = df['Close'].diff()

# Calculate gains (positive values) and losses (negative values)
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)

# Calculate the rolling averages of gains and losses
avg_gain = gain.rolling(window=window_length, min_periods=1).mean()
avg_loss = loss.rolling(window=window_length, min_periods=1).mean()

# Calculate the Relative Strength (RS)
rs = avg_gain / avg_loss

# Calculate RSI
df['RSI'] = 100 - (100 / (1 + rs))

# Check the result
print(df[['Close', 'RSI']].tail())
```
